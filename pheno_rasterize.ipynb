```python
import numpy as np
import pandas as pd
from osgeo import gdal, gdal_array
```

```python
# Define image reading function
def read_image(f):
    ds = gdal.Open(f, gdal.GA_ReadOnly)
    return ds.GetRasterBand(1).ReadAsArray()
```

```python
# read in data
# df = feather.read_dataframe('pheno_nodist_mat.feather')
df = pd.read_csv('../datasets/pheno_nodist_mat.csv', delimiter=",")

```

```python
# open example dataset
ex_img = read_image('../datasets/pheno_example_img.tif')
print(ex_img.shape)

# Get this shape from example raster
#shp = (5126, 3063)
shp = ex_img.shape

# set input dataset for writing raster results
in_ds = gdal.Open('../datasets/pheno_example_img.tif', gdal.GA_ReadOnly)
```

```python
# Get y, x indices from "raveled" index in Eli's results
y, x = np.unravel_index(df.V1.astype(np.int), shp)

for index, year in enumerate(range(1982, 2014)):
    index = index + 10 # indexed on 1
    print('V{index} - {year}'.format(index=index, year=year))
    
    #output raster
    sos = np.zeros(shp)
    
    # Populate with some column
    extract_col = 'V{index}'.format(index=index)
    sos[y, x] = df[extract_col]
    
    # write to raster
    pheno_fn = '../datasets/pheno_013031-014031_spr_{year}.tif'.format(year=year)

    out_driver = gdal.GetDriverByName("GTiff")
    out_ds = out_driver.Create(pheno_fn, 
                               ex_img.shape[1],  # x size
                               ex_img.shape[0],  # y size
                               1,  # number of bands
                               gdal.GDT_Byte)
    out_ds.SetProjection(in_ds.GetProjection())
    out_ds.SetGeoTransform(in_ds.GetGeoTransform())
    out_ds.GetRasterBand(1).WriteArray(sos)
    out_ds.GetRasterBand(1).SetNoDataValue(0)
    #out_ds.GetRasterBand(1).SetColorTable(ct)
    out_ds.GetRasterBand(1).SetDescription('phenology output')
    out_ds = None

    
```
